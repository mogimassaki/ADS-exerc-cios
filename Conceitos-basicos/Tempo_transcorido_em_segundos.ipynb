```python
#Entradas
horas = int(input('Digite o valor da hora: '))
minutos = int(input('Digite o valor do minuto: '))
segundos = int(input('Digite o valor do segundo: '))

#Processamento:
segundos_total = (horas * 3600) + (minutos * 60) + (segundos)

#Saidas:
print(segundos_total)

```
