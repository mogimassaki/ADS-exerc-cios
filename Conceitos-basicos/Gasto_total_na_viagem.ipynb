```python
#Entrada:
distancia = float(input('Digite a distancia a ser percorrida: '))
autonomia = float(input('Digite a autonomia do veiculo: '))
valor_combustivel = float(input('Digite o valor do combustivel: '))

#Processamento:
consumo = distancia / autonomia
gasto = consumo * valor_combustivel

#Saida:
print(f'{gasto:.2f}')
```
