```python
#Entrada:
valor_reais = float(input('Digite o valor em reais: '))
cotacao_dolar = float(input('Digite a cotação do dolar: '))
cotacao_euro = float(input('Digite a cotação do euro: '))
meta_dolar = float(input('Digite a meta em dolar: '))
meta_euro = float(input('Digite a meta em euros: '))

#Processamento:
valor_convertido_dolar = valor_reais / cotacao_dolar
valor_convertido_euro = valor_reais / cotacao_euro
usd_maior_que_eur = valor_convertido_dolar > valor_convertido_euro
atinge_meta_usd = meta_dolar <= valor_convertido_dolar
atinge_meta_eur = meta_euro <= valor_convertido_euro
atinge_ambas_as_metas = atinge_meta_usd and atinge_meta_eur

#Saida:
print(f'USD: {valor_convertido_dolar:.2f}')
print(f'EUR: {valor_convertido_euro:.2f}')
print('usd_maior_que_eur:', usd_maior_que_eur)
print('atinge_meta_usd:', atinge_meta_usd)
print('atinge_meta_eur:', atinge_meta_eur)
print('atinge_ambas_as_metas:', atinge_ambas_as_metas)
```
