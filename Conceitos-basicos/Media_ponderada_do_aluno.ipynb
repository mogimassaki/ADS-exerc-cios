```python
#Entrada:
prova_1 = float(input())
peso_1 = float(input())

prova_2 = float(input())
peso_2 = float(input())

prova_3 = float(input())
peso_3 = float(input())

#Processamento:
media_ponderada = (prova_1 * peso_1 + prova_2 * peso_2 + prova_3 * peso_3) / (peso_1 + peso_2 + peso_3)

#Saida:
print(f'{media_ponderada:.2f}')
```
