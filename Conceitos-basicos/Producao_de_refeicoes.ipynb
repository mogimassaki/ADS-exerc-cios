```python
#Entradas:
numero_viajantes = int(input('Digite o numero de viajantes: '))
numero_dias = int(input('Digite o numero de dias: '))

#Processamento:

total_refeicoes = numero_viajantes * (numero_dias * 3)

#Saidas:
print(total_refeicoes)
```
